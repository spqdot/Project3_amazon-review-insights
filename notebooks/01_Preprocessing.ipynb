### Select the same columns

```python
columns_to_keep = [
    "id",
    "name",
    "brand",
    "categories",
    "asins",
    "reviews.rating",
    "reviews.title",
    "reviews.text",
    "reviews.username",
    "reviews.date"
]

df1 = df1.reindex(columns=columns_to_keep)
df2 = df2.reindex(columns=columns_to_keep)
df3 = df3.reindex(columns=columns_to_keep)
```

### Merge

```python
df = pd.concat([df1, df2, df3], ignore_index=True)
df.head()
```

### Drop Duplicates

```python
df = df.drop_duplicates(
    subset=["id", "reviews.text", "reviews.rating"]
)
```

```python
df.duplicated().sum()
```

### Handle missing values

```python
df = df.dropna(subset=["reviews.rating"])
df = df.dropna(subset=["reviews.text"])
df = df.dropna(subset=["asins"])
```

### Fill review titles

```python
df["reviews.title"] = df["reviews.title"].fillna("No Title")
```

### Fill usernames

```python
df["reviews.username"] = df["reviews.username"].fillna("Unknown")
```

### Fill product names

```python
df["name"] = df.groupby("id")["name"].transform(
    lambda x: x.ffill().bfill()
)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df["name"] = df["name"].fillna("Unknown Product")
```

```python
df["reviews.date"] = pd.to_datetime(
    df["reviews.date"],
    errors="coerce"
)
```
