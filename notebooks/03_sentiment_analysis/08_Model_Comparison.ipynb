### . Introduction

Briefly explain the purpose:

This notebook compares the performance of five sentiment classification models trained on the Amazon Reviews dataset: Logistic Regression, Random Forest, Linear SVM, XGBoost, and BERT Transfer Learning. The comparison is based on accuracy, precision, recall, F1-score, training behavior, and generalization performance.

### Model Comparison Table

```python
import pandas as pd

comparison_df = pd.DataFrame({
    "Model": [
        "Logistic Regression",
        "Random Forest",
        "Linear SVM",
        "XGBoost",
        "BERT"
    ],
    "Train Accuracy": [
        0.8349,
        0.9969,
        0.9133,
        0.9524,
        0.9596
    ],
    "Validation Accuracy": [
        0.8349,
        0.9417,
        0.9142,
        0.9353,
        0.9596
    ],
    "Test Accuracy": [
        0.8349,
        0.9417,
        0.9142,
        0.9353,
        0.9535
    ]
})

comparison_df
```

### Plot the Comparison

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.bar(comparison_df["Model"], comparison_df["Test Accuracy"])

plt.title("Comparison of Test Accuracy Across Models")
plt.xlabel("Models")
plt.ylabel("Accuracy")
plt.ylim(0.75, 1.0)
plt.xticks(rotation=20)

plt.show()
```

### Final Conclusion (Markdown)


Five sentiment analysis models were evaluated on the Amazon Reviews dataset. Logistic Regression provided a simple baseline but achieved the lowest performance. Random Forest significantly improved accuracy but showed signs of overfitting due to its much higher training accuracy than test accuracy. Linear SVM and XGBoost demonstrated strong generalization and competitive performance. BERT Transfer Learning achieved the highest overall performance, with a training accuracy of 95.96%, validation accuracy of 95.96%, and test accuracy of 95.35%. The small difference between training and test accuracy indicates that BERT generalized well without substantial overfitting. Although BERT required greater computational resources and longer training time than the traditional machine learning models, it delivered the best predictive performance by leveraging contextual language understanding.
